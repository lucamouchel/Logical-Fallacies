```python
import pandas as pd 
type = 'cpo'
df = pd.read_json(f'/mnt/lia/scratch/mouchel/Logical-Fallacies/results/llama/arguments/{type}/f-rate.json')
```

```python
df.fallacy_type.value_counts()
test_set = pd.read_json('/mnt/lia/scratch/mouchel/Logical-Fallacies/data/argumentation/test_cckg.json')
```

```python
df2 = df.merge(test_set, on='topic')
```

```python
df2.rename(columns={'argument_y': 'gold', 'argument_x': 'generated'}, inplace=True)
```

```python
data = []

for i, entry in df2.iterrows():
    topic = entry['topic']
    argument = entry['generated']
    stance = 'Supporting' if entry['label'] == 1 else 'Counter'
    predicted_fallacy = ' '.join(word.title() for word in entry['fallacy_type'].split())
    
    data.append({
        'topic': topic,
        'argument': argument,
        'stance': stance,
        'fallacy_type': predicted_fallacy
    })
    
df3 = pd.DataFrame(data)
```
